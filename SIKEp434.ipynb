## Our functions
### $\mathtt{isogeny\_walk(P,E,l,e)}$ for computing the isogeny $\phi:E \to E/\langle P \rangle$ of degree $\ell^e$
### $\mathtt{isogeny\_point(phi,P)}$ for computing $\phi(P)$

```python
#P is point, E is curve, l=2 or 3, l^e is order of P
#returns (phi0, phi1, ..., phi(e-1)) and the final elliptic curve
def isogeny_walk(P,E,l,e):
    phi = []
    Enext = E
    Pnext = P
    for i in range(0,e):
        #final left edge
        Pf = (l^(e-i-1))*Pnext
        #next isogeny
        phi_next = EllipticCurveIsogeny(Enext,Pf)
        Enext = phi_next.codomain()
        Pnext = phi_next(Pnext)
        phi.append(phi_next)
    return (phi,Enext)

#phi is array of isogenies, P: the point
#returns Pnew = phi0 o phi1 o .... o phi(e-1) (P)
def isogeny_point(phi,P):
    Pnext = P
    for cphi in phi:
        Pnext = cphi(Pnext)
    return Pnext
```

# SIKE: Key Generation

```python
# Alice generates her secret/public keys
skA = randint(0, 2^eA-1)
RA = PA + skA * QA
phiA, EA = isogeny_walk(RA, E, 2, eA)
phiA_PB , phiA_QB = isogeny_point(phiA, PB), isogeny_point(phiA, QB) 
pkA = [EA, phiA_PB, phiA_QB]

print "Alice's secret key"
print "skA =", skA, "\n"

print "Alice's public key pkA = [EA, phiA(PB), phiA(QB)]"
print "EA =", EA, "\n"
print "phiA(PB) =", phiA_PB, "\n"
print "phiA(QB) =", phiA_QB
```

# SIKE: Encapsulation

### Input   : Alice's public key pkA
### Output: Ciphertext $c = (c_0, c_1)$ and the shared key $K$ 

```python
# Generate a random string m, and use hash function to compute r = Hash(m || pkA)
m = randomString()
print "m    =", m, "\n"

hh1 = hashlib.sha1()
hh1.update(m + str(pkA[0]) + str(pkA[1]) + str(pkA[2]))
r = ZZ('0x'+hh1.hexdigest()) % (3^eB)
print "r    =", r, "\n"

# Compute c0, i.e pkB 
RB = PB + r * QB
phiB, EB = isogeny_walk(RB, E, 3, eB)
phiB_PA , phiB_QA = isogeny_point(phiB, PA), isogeny_point(phiB, QA) 
pkB = [EB, phiB_PA, phiB_QA]
c0 = pkB
print "c0[0]   =", c0[0], "\n"
print "c0[1]   =", c0[1], "\n"
print "c0[2]   =", c0[2], "\n"



# Compute c1:
# Compute EBA, then j = j_EBA, then h = hash(j), then c1 = h XOR m
RB2 = phiA_PB + r * phiA_QB
phiB2, EBA = isogeny_walk(RB2, EA, 3, eB)

j = EBA.j_invariant()

hh2 = hashlib.sha1()
hh2.update(str(j))

c1 = hex(ZZ('0x'+hh2.hexdigest()) ^^ ZZ('0x'+m.encode('hex')))
print "c1   =", c1, "\n"

# Shared Secret K = hash(m || c0 || c1)
K = hashlib.sha1()
K.update(m + str(c0) + c1)
K_Bob = K.hexdigest()
print "K_Bob =", K_Bob, "\n"
```
